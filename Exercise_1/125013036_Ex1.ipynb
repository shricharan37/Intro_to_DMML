```python
#Question1
import numpy as np
import pandas as pd
import statistics as stat
import matplotlib.pyplot as plt
Data = 15,20, 30,70, 80, 50, 20, 15, 10, 10, 15, 8, 12, 18, 14, 20, 22, 25, 16, 10, 14, 18, 22, 26,30
print(Data)
print("Median is",np.median(Data))
print("Mean is",np.mean(Data))
print("Mode is",stat.mode(Data))
print("Variance is",np.var(Data))
print("Standard deviation is",np.std(Data))
```

```python
#Question7
#Upset_plot
import pandas as pd
from upsetplot import plot
biological_data_A ={'Gene1', 'Gene2', 'Gene3', 'Gene4', 'Gene5'}
biological_data_B ={'Gene3', 'Gene4', 'Gene5', 'Gene6', 'Gene7'}
set_names = ['biological_data_A', 'biological_data_B']
all_elems = biological_data_A.union(biological_data_B)
df = pd.DataFrame([[e in biological_data_A, e in biological_data_B] for e in all_elems], columns = set_names)
df_up = df.groupby(set_names).size()
plot(df_up, orientation='horizontal')
```

```python
#Question8
#Sankey_plot
import plotly.graph_objects as go
nodes = ["Species A", "Species B", "Species C", "Gene A1", "Gene A2", "Gene B1", "Gene B2", "Gene C1", "Gene C2"]
links = [
    {"source": "Gene B1", "target": "Species A", "value": 1},
    {"source": "Gene B1", "target": "Species B", "value": 1},
    {"source": "Gene A2", "target": "Species C", "value": 1},
    {"source": "Gene A1", "target": "Species B", "value": 1},
    {"source": "Gene A2", "target": "Species A", "value": 1},
    {"source": "Gene A1", "target": "Species C", "value": 1},
    {"source": "Gene B2", "target": "Species A", "value": 1},
    {"source": "Gene B2", "target": "Species C", "value": 1},
    {"source": "Gene C2", "target": "Species A", "value": 1},
    {"source": "Gene C2", "target": "Species B", "value": 1},
    {"source": "Gene C1", "target": "Species C", "value": 1},
    {"source": "Gene C2", "target": "Species B", "value": 1},
]

# Create Sankey diagram
fig = go.Figure(data=[go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=nodes
    ),
    link=dict(
        source=[nodes.index(link["source"]) for link in links],
        target=[nodes.index(link["target"]) for link in links],
        value=[link["value"] for link in links]
    )
)])

# Customize layout
fig.update_layout(title_text="Sankey Diagram of Gene Flow to Species",
                  font_size=10)

# Show plot
fig.show()
```
