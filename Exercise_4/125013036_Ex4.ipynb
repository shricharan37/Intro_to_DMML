```python
#Question2
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler

iris = load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
print("Before normalization:")
print(data.head())
scaler = MinMaxScaler()
normalized = scaler.fit_transform(data)
normalized_data = pd.DataFrame(data=normalized, columns=iris.feature_names)
print("\nAfter normalization:")
print(normalized_data.head())
```

```python
#Question3
from scipy.stats import zscore
print("Before standardization:")
print(data.head())
standardized_ = zscore(data)
standardized_data1 = pd.DataFrame(data=standardized, columns=iris.feature_names)
print("\nAfter standardization:")
print(standardized_data1.head())
```

```python
#Question4
from sklearn.preprocessing import FunctionTransformer

print("Before transformation:")
print(data.head())

log_transform_func = FunctionTransformer(np.log1p, validate=True)
sqrt_transform_func = FunctionTransformer(np.sqrt, validate=True)
boxcox_transform_func = FunctionTransformer(lambda x: np.power(x, 0.5), validate=True)

log_transformed = log_transform_func.transform(data)

sqrt_transformed = sqrt_transform_func.transform(data)

boxcox_transformed = boxcox_transform_func.transform(data)

log_transformed_data = pd.DataFrame(data=log_transformed, columns=iris.feature_names)
sqrt_transformed_data = pd.DataFrame(data=sqrt_transformed, columns=iris.feature_names)
boxcox_transformed_data = pd.DataFrame(data=boxcox_transformed, columns=iris.feature_names)

print("After transformation - Log Transform:")
print(log_transformed_data.head())

print("\nAfter transformation - Square Root Transform:")
print(sqrt_transformed_data.head())

print("\nAfter transformation - Box-Cox Transform:")
print(boxcox_transformed_data.head())
```

```python
from sklearn.preprocessing import KBinsDiscretizer
print("Before discretization:")
print(data.head())

bins = 5

equal_width_discretizer = KBinsDiscretizer(n_bins=bins, encode='ordinal', strategy='uniform')
equal_width_discretized = equal_width_discretizer.fit_transform(data)

equal_depth_discretizer = KBinsDiscretizer(n_bins=bins, encode='ordinal', strategy='quantile')
equal_depth_discretized = equal_depth_discretizer.fit_transform(data)

equal_width_discretized_df = pd.DataFrame(data=equal_width_discretized, columns=iris.feature_names)
equal_depth_discretized_df = pd.DataFrame(data=equal_depth_discretized, columns=iris.feature_names)

print("\nAfter discretization - Equal-Width Binning:")
print(equal_width_discretized_df.head())

print("\nAfter discretization - Equal-Depth Binning:")
print(equal_depth_discretized_df.head())

#Interpretation
#Equal-Width Binning: The data has been divided into bins of equal width.
#The data has been divided into bins of equal depth. Each bin has the same number of datapoints
```
