```python
import numpy as np
from scipy.stats import norm, chi2

def detect_univariate_outliers_max_likelihood(data, threshold=0.01):
    mean, std = norm.fit(data)
    likelihoods = norm.pdf(data, mean, std)
    return np.where(likelihoods < threshold)[0]

def detect_univariate_outliers_z_score(data, threshold=3):
    mean, std = np.mean(data), np.std(data)
    z_scores = np.abs((data - mean) / std)
    return np.where(z_scores > threshold)[0]

def detect_multivariate_outliers(data, threshold=0.05):
    mean = np.mean(data, axis=0)
    cov_matrix = np.cov(data, rowvar=False)
    diff = data - mean
    mahalanobis_distances = np.sqrt(np.sum((np.dot(diff, np.linalg.inv(cov_matrix)) * diff), axis=1))
    chi_sq_stats = np.sum((np.dot(diff, np.linalg.inv(cov_matrix)) * diff), axis=1)
    md_threshold = chi2.ppf(1 - threshold, data.shape[1])
    return np.where((mahalanobis_distances > md_threshold) | (chi_sq_stats > chi2.ppf(1 - threshold, data.shape[1])))[0]

def remove_outliers(data, outliers):
    return data[~np.isin(np.arange(len(data)), outliers)]

def evaluate_effectiveness(data, outliers):
    ml_outliers = detect_univariate_outliers_max_likelihood(data)
    z_scores = detect_univariate_outliers_z_score(data)
    md_chi_sq = detect_multivariate_outliers(data)

    removed_ml_outliers = remove_outliers(data, outliers[ml_outliers])
    removed_z_scores = remove_outliers(data, outliers[z_scores])
    removed_md_chi_sq = remove_outliers(data, outliers[md_chi_sq])

    return len(removed_ml_outliers), len(removed_z_scores), len(removed_md_chi_sq)

data = np.random.normal(loc=0, scale=1, size=(1000, 2))
data[5] = [10, 10]
data[15] = [-10, -10]

outliers = np.arange(len(data))
ml_effectiveness, z_score_effectiveness, md_chi_sq_effectiveness = evaluate_effectiveness(data, outliers)
print("Maximum Likelihood effectiveness:", ml_effectiveness)
print("Z-score effectiveness:", z_score_effectiveness)
print("Mahalanobis Distance and Chi-square Statistics effectiveness:", md_chi_sq_effectiveness)
```
