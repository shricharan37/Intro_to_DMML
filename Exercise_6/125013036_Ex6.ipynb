```python
import pandas as pd
data = pd.read_csv("breast-cancer-wisconsin.csv")
data.replace('?', np.nan, inplace=True)
data.dropna(inplace=True)
```

```python
dataset = pd.read_csv("breast-cancer-wisconsin.csv")

dependent_variable = data.pop("Class")
independent_variables = data

correlations = independent_variables.corrwith(dependent_variable)

threshold = 0.6

significant_variables = correlations[abs(correlations) > threshold].index

selected_data = dataset[significant_variables]

print(significant_variables)
```

```python
selected_data
```

```python
from scipy import stats
min_rows = min(selected_data.shape[0], dependent_variable.shape[0])
selected_data = selected_data[:min_rows]
dependent_variable = dependent_variable[:min_rows]


combined_dataset = np.column_stack((selected_data, dependent_variable))

z_scores_combined = np.abs(stats.zscore(combined_dataset))

threshold = 3

cleaned_dataset = combined_dataset[(z_scores_combined < threshold).all(axis=1)]

cleaned_selected_data = cleaned_dataset[:, :-1]
cleaned_dependent_variable = cleaned_dataset[:, -1]
```

```python
cleaned_selected_data
```

```python
cleaned_dependent_variable
```

```python
from sklearn.preprocessing import StandardScaler
scaler_standardization = StandardScaler()
selected_data_standardized = scaler_standardization.fit_transform(cleaned_selected_data)
selected_data_standardized
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(selected_data_standardized,cleaned_dependent_variable, test_size=0.2, random_state=42)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

print("Classification Report:")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
