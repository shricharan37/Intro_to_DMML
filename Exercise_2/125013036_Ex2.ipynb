```python
data= data.drop(columns = ["ID","CommonName","MfgName","ScientificName","FoodGroup","ShortDescrip","Descrip"])
```

```python
scaler=StandardScaler()
scaler.fit(data)
data_scaled=scaler.transform(data)
data_scaled
```

```python
pca=PCA(n_components=4)
pca.fit(data_scaled)
data_pca=pca.transform(data_scaled)
data_pca
```

```python
def biplot(score,coef,label=None):
  xs=score[:,0]
  ys=score[:,1]
  n=coef.shape[0]
  scalex=1.0/(xs.max()-xs.min())
  scaley=1.0/(ys.max()-ys.min())
  plt.scatter(xs * scalex, ys * scaley, s=5, color="green")
  for i in range(n):
    plt.arrow(0,0, coef[i,0], coef[i,1], color="blue", alpha=0.5)
    plt.text(coef[i,0]*1.15, coef[i,1]*1.15, label[i], color='black', ha='center', va="center")

  plt.xlabel("PC{}".format(1))
  plt.ylabel("PC{}".format(2))
  plt.figure()

plt.figure(figsize=(15,30))
plt.title("Biplot")
biplot(data_pca, np.transpose(pca.components_), list(data.columns))
plt.show
```
