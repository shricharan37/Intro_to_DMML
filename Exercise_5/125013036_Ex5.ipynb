```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
dataset = pd.read_csv('kc_house_data.csv')
dataset
```

```python
dataset = pd.read_csv("kc_house_data.csv")

dependent_variable = dataset.pop("price")
independent_variables = dataset

correlations = independent_variables.corrwith(dependent_variable)

threshold = 0.6

significant_variables = correlations[abs(correlations) > threshold].index

selected_data = dataset[significant_variables]

print(significant_variables)
```

```python
selected_data
```

```python
print(dataset.isnull().sum())
```

```python
from scipy import stats

combined_dataset = np.column_stack((selected_data,dependent_variable))

z_scores_combined = np.abs(stats.zscore(combined_dataset))

threshold = 3

cleaned_dataset = combined_dataset[(z_scores_combined < threshold).all(axis=1)]

cleaned_selected_data = cleaned_dataset[:, :-1]
cleaned_dependent_variable = cleaned_dataset[:, -1]
```

```python
cleaned_selected_data
```

```python
cleaned_dependent_variable
```

```python
from sklearn.preprocessing import StandardScaler
scaler_standardization = StandardScaler()
selected_data_standardized = scaler_standardization.fit_transform(cleaned_selected_data)
selected_data_standardized
```

```python
from sklearn.preprocessing import StandardScaler
scaler_standardization_dep = StandardScaler()
dependent_variable_standardized = scaler_standardization_dep.fit_transform(cleaned_dependent_variable.reshape(-1, 1))
dependent_variable_standardized
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(selected_data_standardized, dependent_variable_standardized, test_size=0.2, random_state=42)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

```python
y_test_1d = y_test.ravel()
y_pred_1d = y_pred.ravel()

reg_model_diff = pd.DataFrame({'Actual value': y_test_1d, 'Predicted value': y_pred_1d})
print(reg_model_diff)
```

```python
from sklearn import metrics
mae = metrics.mean_absolute_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
r2 = np.sqrt(metrics.mean_squared_error(y_test, y_pred))

print('Mean Absolute Error:', mae)
print('Mean Square Error:', mse)
print('Root Mean Square Error:', r2)
```
